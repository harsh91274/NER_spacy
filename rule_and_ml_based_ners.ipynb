```python
import json

with open ("J. K. Rowling - Harry Potter 1 - Sorcerer's Stone.txt", "r", encoding="utf-8") as f:
    text=f.read().split("\n\n")
    print(text)
```

```python
character_names=[]
with open("hp_characters.json", "r", encoding="utf-8") as f:
    characters=json.load(f)

    for character in characters:
        names=character.split()
        for name in names:
            if "and"!=name and "the"!=name and "The"!=name:
                name=name.replace(",", "").strip()
                character_names.append(name)
```

```python
for segment in text:
    #print(segment)
    #remove whitepsace 
    segment.strip()
    #remove linebreaks
    segment=segment.replace("\n", " ")
    #print(segment)
    #remove punctuations
    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~''' 
    for ele in segment:
        if ele in punc:
            segment = segment.replace(ele, "")
    print(segment)

    words=segment.split()
    #print(words)

    i=0
    for word in words:
        if word in character_names:
            if words[i-1][0].isupper():
                print(f"Found Characters: {words[i-1]} {word}")
            else:
                print(f"Found Characters: {word}")
        
        i=i+1
```

#### Lot of code for a simple entity recognition! ML based NER's are much more versatile and efficient

```python
import spacy
test=text[4:7][0]
print(test)
```

```python
nlp=spacy.load("en_core_web_lg")
```

```python
doc=nlp(test)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

Dursleys is incorrectly classified as an ORG. This is because this is a general purpose model and not fine tuned to a specific domain or dataset
